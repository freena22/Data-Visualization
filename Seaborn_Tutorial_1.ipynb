## Part I: Controlling figure aesthetics

- Seaborn can acheived a more customizable and attractive plot than Matplotlib. 
- Seaborn comes with a number of customized themes and a high-level interface for controlling the look of matplotlib figures.

### Matplot vs. Seaborn in aesthetics tone

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))
```

- define a simple function to plot some offset sine waves

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip) #
```

```python
sinplot()  # matplotlib defaults
```

```python
import seaborn as sns
sinplot()
```

### Styling figures with axes_style() and set_style()

- There are five preset seaborn themes: darkgrid(default), whitegrid, dark, white, and ticks. 
-  the grid helps the plot serve as a lookup table for quantitative information, and the white-on grey helps to keep the grid from competing with lines that represent data. The whitegrid theme is similar, but it is better suited to plots with heavy data elements.

```python
sns.set_style("ticks")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style('dark')
sinplot()
```

### Removing spines with despine()

- Call the seaborn function despine() to remove the top and right axes spines

```python
sns.set_style("white")
sinplot() # default as no parameters -- delete the uppper and right 
sns.despine()
```

```python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)
# offset the distance between x and y
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True) # delete the left
st = sns.axes_style("darkgrid")
```

- Remove the top and right: despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)

### Temporarily setting figure style

- you can also use the axes_style() function in a with statement to temporarily set plot parameters

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

### Overriding elements of the seaborn styles

- If you want to customize the seaborn styles, you can pass a dictionary of parameters to the rc argument of axes_style() and set_style(). 

```python
sns.axes_style()
```

```python
# Try to set different versions of these parameters:
sns.set_style("darkgrid", {'grid.color': "pink"})
sinplot()
```

### Scaling plot elements with plotting_context() and set_context()

- The four preset contexts, in order of relative size, are paper, notebook, talk, and poster. The notebook style is the default, and was used in the plots above.

```python
sns.set()
```

```python
sns.set_context('paper')
plt.figure(figsize=(8,6))
sinplot()
```

```python
sns.set_context('poster')
plt.figure(figsize=(8,6))
sinplot()
```

-  Call set_context() with one of these names to set the parameters, and you can override the parameters by providing a dictionary of parameter values.

```python
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 1.5})
sinplot()
```

- Both the style and the context can be quickly configured with the set() function.
